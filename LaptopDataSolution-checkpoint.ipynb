**Extracting data from file. **

```python
import pandas as pd
df=pd.read_csv("Laptop.csv")
df
```

######  Count of each laptop according their brand in given dataset.

```python
brand_counts = df['Company'].value_counts()
print(brand_counts)
```

###### Bar graph which represent Operating system.

```python
 import matplotlib.pyplot as plt
osc = df['OpSys'].value_counts()
osc.plot(kind='bar', color='skyblue')
plt.title('Count of Laptops by Operating System')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--')
plt.show()
```

###### create a pie dig according to laptop brands names. 

```python

import matplotlib.pyplot as plt
br = df['Company'].value_counts()
br.plot(kind='pie',autopct='%1.1f%%')
plt.title('laptop distribution')
plt.show()
```

###### wordcloud with company names.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from collections import Counter

# Extract company names and convert to lowercase
comp = df['Company'].str.lower().tolist()

# Create a Counter object from the list of company name
word_cloud_list = Counter(comp)

# Generate Word Cloud
wordcloud = WordCloud(
    width=3000,
    height=2000,
    background_color="white",
    collocations=False,
    stopwords=STOPWORDS,
).generate_from_frequencies(word_cloud_list)

# Display the generated word cloud
plt.figure(figsize=(15, 8))
plt.title('Various Company in Laptop', fontsize=20, fontweight='bold')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
